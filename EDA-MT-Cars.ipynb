```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_excel('mtcars.xlsx')
cars.head()
```

```python
cars.shape
```

```python
cars.info()
```

```python
cars.describe()
```

```python
# Check null values
cars.isna().sum()
```

```python
# Replacing the missing values in weight column by its average weight
cars.wt.fillna(np.mean(cars['wt']), inplace = True)
```

```python
cars.gear.value_counts()
```

```python
cars.gear.fillna(3.0,inplace = True)
```

```python
cars.isna().sum()
```

```python
# checking for duplicate values
cars.duplicated().sum()
```

```python
cars.drop_duplicates()
```

```python
# Distribution of cars wrt cylinders
cylinders = cars.cyl.value_counts()
cylinders
```

```python
plt.pie(cylinders.values, labels  = ['8 Cyl', '4 Cyl', '6 Cyl'], autopct="%1.1f%%",
       colors=['steelblue','lightgreen','salmon'])

plt.show()
```

```python
# Finding distribution of am column
gt = cars.am.value_counts()
gt
```

```python
plt.pie(gt.values, labels  = ['Automatic', 'Manual'], autopct="%1.1f%%",
       colors=['steelblue','lightgreen'])

plt.show()
```

```python
# Finding distribution of gear column
gears = cars.gear.value_counts()
gears
```

```python
plt.bar(gears.index,gears.values,color = 'salmon')
plt.xlabel('Number of Gears')
plt.ylabel('Count of Cars')
plt.show()
```

```python
gears.values
```

```python
# Finding relationship of wt and mileage
plt.scatter(cars.wt, cars.mpg_high, color = 'steelblue')
plt.xlabel('Weight of cars')
plt.ylabel('Mileage of car on Highway')
plt.show()
```

```python
# Finding relationship of HP and displacement on city
plt.scatter(cars.disp, cars.mpg_high, color = 'lightgreen')
plt.xlabel('Displacement of Car')
plt.ylabel('Mileage of car')
plt.show()
```

```python
# Frequency distribution
plt.hist(cars['wt'], bins = 10)
plt.show()
```

### Above graph shows the fact that the most of the cars have their weights in range of 3 to 3.8 tonns

```python
# Find the cars whose mileage is greater than 20
cars[cars['mpg_high']>22][['model','mpg_high']]
```

```python
# Find the cars whose city mileage is greater than 16 and highway mileage greater than 19
cars[(cars['mpg_city']>16) & (cars['mpg_high']>19)][['model','mpg_high','mpg_city']]
```

```python
# Find all cars with 5 gear
cars[cars['gear']==5.0]
```

```python
# Find all cars with 5 gear and 6 cylinders
cars[(cars['gear']==5.0) & (cars['cyl']==6)]
```

```python
# Find cars whose highway mileage is greater than average highway mileage
avg_mil_high = np.mean(cars['mpg_high'])
cars[cars['mpg_high']>avg_mil_high][['model','mpg_high']]
```
